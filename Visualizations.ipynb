**SET UP YOUR KITCHEN**

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

url = 'https://raw.githubusercontent.com/ryanecole2-lang/Machine-Learning-With-Python-Bootcamp/refs/heads/main/gari_price_data.csv'
url_2 = 'https://raw.githubusercontent.com/ryanecole2-lang/Machine-Learning-With-Python-Bootcamp/refs/heads/main/cassava_variety%20(1).csv'

# load datasets
gari = pd.read_csv(url)
cassava = pd.read_csv(url_2)

#Global style - Clean white background for projector
sns.set_theme(style='whitegrid', palette='muted', font_scale=1.1)
mpl.rcParams['figure.dpi'] = 120
print(gari.shape, cassava.shape)
```

**Line Plot**

```python
from os import set_inheritable
fig, ax = plt.subplots(figsize=(8, 5))
day = [1,2,3,4,5] # Adjusted to match the length of price
price = [110, 220, 330, 440, 550]
ax.plot(day, price, color='#3B3F8C', linewidth=2.2, marker='o', label='Gari price (LRD)')
ax.set_title('Price Over Weeks - Gobachop')

ax.set_xlabel('Days')
ax.set_ylabel('Price (LRD)')
ax.legend()
plt.show()
```

**SCATTER PLOT: PRICE VS RAINFALL**

```python
fig, ax = plt.subplots(figsize=(8, 5))
sns.regplot(data=gari, x='rainfall_mm',
y='price_next_day_lrd', ax=ax,
scatter_kws={'color': '#008080'},
line_kws={'color': '#D97706'})
ax.set_title('Price vs Rainfall _Red Light')
plt.show()
```

**BAR CHART: AVERAGE PRODUCT**

```python
import matplotlib.pyplot as plt

#1. Create a dummy data for our two market products
product = ['Gari Bag', 'Cassava Bag']
prices = [450, 300]

#2. Plot the data, add labels, and display
bars = plt.bar(product, prices, color=['#3B3F8C', '#008080'])
plt.bar_label(bars, fmt='LRD %. 0f', padding=3)
plt.title('Market Prices - Gobachop')
plt.show()
```

**HISTOGRAM: TRANSPORT COST**

```python
import matplotlib.pyplot as plt

#1. Dummy transport fares paid by 7 different market women
fares = [150, 150, 150, 300, 300, 500, 500]

#2. Group the fares into matching "buckets" and display
plt.hist(fares, bins=[100, 200, 400, 600], color='#7C3AED', edgecolor='white')
plt.title('Transport cost Distribution')
plt.xlabel('Fare Price (LRD)')
plt.ylabel('Number of Women')
plt.show()
```

**BAR PLOT: SPOTTING BAD DATA**

```python
import matplotlib.pyplot as plt

#1. Normal price (around 150-200) + one "bad data" mistake (2000 LRD!)
gari_prices = [150, 160, 170, 180, 200, 210, 2000]

#2. Draw the boxplot to catch the error
plt.boxplot(gari_prices, patch_artist=True, boxprops=dict(facecolor='#3B3F8C'))
plt.title('spotting Bad Data (Gari Price)')
plt.ylabel('Price (LRD)')
plt.show()

```

**HEATMAP: SHOWING FEATURE CORRELATION**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create meaningful data with a mix of correlations
data = {
    'Rainfall': [100, 80, 60, 40, 20],  # decreases
    'Temperature': [22, 24, 26, 28, 30], # increases  negative with Rainfall
    'Crop_Yield': [5.0, 6.2, 7.1, 7.5, 7.0], # rises then falls (non-linear effect)
    'Fertilizer': [10, 20, 30, 40, 50], # increase  positive with yield partly
    'Pest_Damage': [30, 25, 18, 20, 35] # U-shape  weak/negative with others
}
df = pd.DataFrame(data)

# Compute correlation matrix
corr = df.corr()

# Set up a larger, cleaner figure
plt.figure(figsize=(8, 6))

# Draw heatmap with full colour range (-1 to 1)

sns.heatmap(corr,
            annot=True,    # show white at 0
            cmap='coolwarm', # red = negative, white = 0, blue = positive
vmin=-1, vmax=1,
center=0,    # ensure white at 0
square=True,  # square cells
linewidths=.5, # thin grid lines
annot_kws={'size': 10})
plt.title('Feature Correlation Matrix - A Beautiful Mix of Relationships', fontsize=14)
plt.tight_layout()
plt.show()
```
